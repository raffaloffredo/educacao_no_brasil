```python
# importar bibliotecas
import matplotlib.pyplot as plt
from pywaffle import Waffle
```

# Configuração de Cores

```python
azul_escuro = '#004A8F'
cinza_escuro = '#424040'
cinza_claro = '#BFBFBF'
azul_claro = '#91AAC9'
```

# População Total Projetada 2023

```python
# população total

# cada quadrado representa 1 milhão de pessoas
# total projetado: 216.284.269

fig = plt.figure(
                FigureClass=Waffle,
                figsize=(7, 5),
                rows=10,
                columns=22,
                values={'População Brasileira': 100},
                colors=[azul_claro],
                legend={'loc': 'upper left',
                        'bbox_to_anchor': (1.1, 1),
                        'frameon': False,
                        'labelcolor': '#424040'},
                )

# título do gráfico
plt.text(0, 1.2, 'População Total Projetada para 2023',
         fontsize = 18,
         weight = 'bold',
         color=azul_escuro
         )

# subtítulo
plt.text(0, 1.15, '216.284.269 pessoas',
         verticalalignment = 'top',
         fontsize = 12,
         color=cinza_escuro
         )

# rodapé
plt.text(0, -0.1, 'Fonte: IGBE',
         verticalalignment = 'bottom',
         fontsize = 8,
         color=cinza_escuro
         )

fig.set_tight_layout(False)
plt.show()
```

# Acima de 15 anos: Alfabetizados x Analfabetos

```python
# acima de 15 anos
# analfabetos x alfabetizados

# cada quadrado representa 1 milhão de pessoas
# total de pessoas acima de 15 anos: 172.431.571
# população abaixo de 15 anos: 43.852.698
# alfabetizados: 161.051.087
# analfabetos: 6.6% ou 11.380.484

data = {'Analfabetos': 11380484,
        'Alfabetizados': 161051087}

fig = plt.figure(
                FigureClass=Waffle,
                figsize=(7, 5),
                rows=10,
                columns=16,
                values=data,
                colors=[cinza_claro, azul_escuro],
                legend={'loc': 'upper left',
                        'bbox_to_anchor': (1.1, 1),
                        'frameon': False,
                        'labelcolor': '#424040'}
                )

# título do gráfico
plt.text(0, 1.2, 'Analfabetos x Alfabetizados',
         fontsize = 18,
         weight = 'bold',
         color=azul_escuro
         )

# subtítulo
plt.text(0, 1.15, '93.4% da população com idade acima de 15 anos é alfabetizada',
         verticalalignment = 'top',
         fontsize = 12,
         color=cinza_escuro
         )

# rodapé
plt.text(0, -0.1, 'Fonte: IGBE',
         verticalalignment = 'bottom',
         fontsize = 8,
         color=cinza_escuro
         )

fig.set_tight_layout(False)
plt.show()
```

# Acima de 25 +  Ensino médio

```python
# acima de 25 anos
# concluiu o ensino médio

# cada quadrado representa 1 milhão de pessoas
# total de pessoas ABAIXO de 25 anos: 75.337.182
# total de pessoas ACIMA de 25 anos: 140.947.087
# completou ensino médio: 48.8% ou 68.782.178
# não completou o ensino médio: 51.2% ou 72.164.909

data={'Não completou': 51.2,
      'Ensino médio completo': 48.8}

fig = plt.figure(
                FigureClass=Waffle,
                rows=10,
                columns=14,
                values=data,
                colors=[cinza_claro, azul_escuro],
                legend={'loc': 'upper left',
                        'bbox_to_anchor': (1.1, 1),
                        'frameon': False,
                        'labelcolor': '#424040'}
                )

# título do gráfico
plt.text(0, 1.25, 'Adultos que completaram o Ensino Médio',
         fontsize = 18,
         weight = 'bold',
         color=azul_escuro
         )

# subtítulo
plt.text(0, 1.20, 'Apenas 48.8% da população com idade acima de 25 anos possui o\nensino médio completo',
         verticalalignment = 'top',
         fontsize = 12,
         color=cinza_escuro
         )

# rodapé
plt.text(0, -0.1, 'Fonte: IGBE',
         verticalalignment = 'bottom',
         fontsize = 8,
         color=cinza_escuro
         )

fig.set_tight_layout(False)
plt.show()
```

# Acima de 25 anos + ensino superior

```python
# acima de 25 anos
# concluiu o ensino superior

# cada quadrado representa 1 milhão de pessoas
# total de pessoas ABAIXO de 25 anos: 75.337.182
# total de pessoas acima de 25 anos: 140.947.087
# completou ensino superior: 24.524.793
# não completou o ensino superior: 116.422.294

fig = plt.figure(
                FigureClass=Waffle,
                rows=10,
                columns=14,
                values={'Não concluiu': 82.6,
                        'Ensino superior completo': 17.4},
                colors=[cinza_claro, azul_escuro],
                legend={'loc': 'upper left',
                        'bbox_to_anchor': (1.1, 1),
                        'frameon': False,
                        'labelcolor': '#424040'}
                )

# título do gráfico
plt.text(0, 1.25, 'Adultos com Ensino Superior completo',
         fontsize = 18,
         weight = 'bold',
         color=azul_escuro
         )

# subtítulo
plt.text(0, 1.20, 'Apenas 17.4% da população, acima de 25 anos completou o\nensino superior',
         verticalalignment = 'top',
         fontsize = 12,
         color=cinza_escuro
         )

# rodapé
plt.text(0, -0.1, 'Fonte: IGBE',
         verticalalignment = 'bottom',
         fontsize = 8,
         color=cinza_escuro
         )

fig.set_tight_layout(False)
plt.show()
```

# Proficiente

```python
# proficiente

# entre 15 e 64 anos = 148.985.045
# fora da faixa (abaixo de 25 anos)= 67.299.224
# 12% da população entre 15 e 64 anos = 17.878.205
# não proficiente: 131.106.840

fig = plt.figure(
                FigureClass=Waffle,
                rows=10,
                columns=15,
                values={'Não proficiente': 88.46,
                        'Proficiente': 11.54},
                colors=[cinza_claro, azul_escuro],
                legend={'loc': 'upper left',
                        'bbox_to_anchor': (1.1, 1),
                        'frameon': False,
                        'labelcolor': '#424040'}
                )

# título do gráfico
plt.text(0, 1.2, 'População proficiente',
         fontsize = 18,
         weight = 'bold',
         color=azul_escuro
         )

# subtítulo
plt.text(0, 1.15, 'Entre 25 e 64 anos 11.54% da população é considerada proficiente',
         verticalalignment = 'top',
         fontsize = 12,
         color=cinza_escuro
         )

# rodapé
plt.text(0, -0.1, 'Fonte: IBGE, INAF',
         verticalalignment = 'bottom',
         fontsize = 8,
         color=cinza_escuro
         )

fig.set_tight_layout(False)
plt.show()
```

```python
# proficiente

# população em 2023	= 216.284.269
# entre 25 e 59 anos = 107.267.101
# com ensino superior = 18.664.475
# proficiente = 6.142.340

# restante: 210.141.929

fig = plt.figure(
                FigureClass=Waffle,
                rows=10,
                columns=22,
                values={'População Restante': 210141929,
                        'População Diferenciada': 6142340},
                colors=[cinza_claro, azul_escuro],
                legend={'loc': 'upper left',
                        'bbox_to_anchor': (1.1, 1),
                        'frameon': False,
                        'labelcolor': '#424040'}
                )

# título do gráfico
plt.text(0, 1.30, 'População brasileira x pessoas proficientes',
         fontsize = 18,
         weight = 'bold',
         color=azul_escuro
         )

# subtítulo
plt.text(0, 1.25, 'Do total da população menos de 3% possui o Ensino Superior completo e\né considerada proficiente na língua portuguesa.',
         verticalalignment = 'top',
         fontsize = 12,
         color=cinza_escuro
         )

# rodapé
plt.text(0, -0.2, 'Essa porcentagem refere-se à pessoas na faixa etária entre 25 e 59 anos.\nFonte: IGBE, INAF',
         verticalalignment = 'bottom',
         fontsize = 8,
         color=cinza_escuro
         )

fig.set_tight_layout(False)
plt.show()
```
